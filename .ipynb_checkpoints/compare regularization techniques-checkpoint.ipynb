# Compare regularization techniques

```python
import numpy as np

import cPickle as pickle
import matplotlib.pyplot as plt
import matplotlib.cm as cm
%matplotlib inline
```

## Early stages of learning

```python
reg_types = {'noreg':'r-', 'norm':'b-', 'mag':'g-', 'drop':'c-', 'gauss':'k-'}
handles = []
for reg_type in reg_types.keys():
    with open(reg_type + '.p', 'r') as f:
        acc_curve = pickle.load(f)

    plt.figure(0, figsize=(8,4))
    plt.title("Model accuracy")

    stop = len(acc_curve)/5
    h, = plt.plot(acc_curve[:stop,0], acc_curve[:stop,1], reg_types[reg_type], label=reg_type)
    handles.append(h)
    
plt.legend(handles=handles, loc=4)
```

## Convergence behaviors

```python
handles = []
for reg_type in reg_types.keys():
    with open(reg_type + '.p', 'r') as f:
        acc_curve = pickle.load(f)

    plt.figure(0, figsize=(8,4))
    plt.title("Model accuracy")

    start = len(acc_curve)/5
    h, = plt.plot(acc_curve[start:,0], acc_curve[start:,1], reg_types[reg_type], label=reg_type)
    handles.append(h)
    
plt.legend(handles=handles, loc=4)
```

## Full training data

```python
handles = []
for reg_type in reg_types.keys():
    with open(reg_type + '.p', 'r') as f:
        acc_curve = pickle.load(f)

    plt.figure(0, figsize=(8,4))
    plt.title("Model accuracy")

    h, = plt.plot(acc_curve[1:,0], acc_curve[1:,1], reg_types[reg_type], label=reg_type)
    handles.append(h)
    
plt.legend(handles=handles, loc=4)
```
